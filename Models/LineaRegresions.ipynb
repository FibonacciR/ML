### Data Generation
We use artifiacially hgenerated data to demonstrate the linear regression model. The data is generated using the following equation:

```python
# Step 1: Import necessary libraries
import numpy as np
import pandas as pd

# Set random seed for reproducibility
np.random.seed(42)
n_samples = 100  # Number of samples

# Generate an independent variable (X) and a dependent variable (y) with a linear relationship
X = 2 * np.random.rand(n_samples, 1)
y = 4 + 3 * X + np.random.randn(n_samples, 1) * 0.5  # y = 4 + 3X + noise

# Convert to a DataFrame for easier analysis
data = pd.DataFrame(np.hstack([X, y]), columns=['X', 'y'])

# Display the first few rows of the generated data
data.head()

# git remote add origin https://github.com/https://github.com/FibonacciR/ML.git
```

### Data Visualization

```python
# Step 2: Visualize the generated data
import matplotlib.pyplot as plt

# Plot the data points
plt.figure(figsize=(8, 6))
plt.scatter(data['X'], data['y'], color='blue', alpha=0.6)
plt.xlabel('Independent Variable (X)')
plt.ylabel('Dependent Variable (y)')
plt.title('Generated Data for Linear Regression')
plt.show()
```

### Model Fitting
Here we fit a linear regression model to the generated data. The model will learn the relationship between the independent variable (X) and the dependent variable (y) using the training data.

This is a simple linear regression problem, where the model learns the relationship between X and y using a linear equation of the form:
y = mx + b
where:
y is the dependent variable,
x is the independent variable,
m is the slope of the line,
b is the y-intercept.
The goal of the model is to find the best values for the slope and the intercept that minimize the difference between the predicted values and the actual values of y in the training data.

```python
# Step 3: Model fitting
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(data[['X']], data['y'], test_size=0.2, random_state=42)

# Initialize and train the Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train) 

# git commit -m "Models Uploaded"


# Model parameters
intercept = model.intercept_
slope = model.coef_[0]

print(f"Intercept: {intercept}")
print(f"Slope: {slope}")

# Plot the data and the fitted regression line
plt.figure(figsize=(8, 6))
plt.scatter(data['X'], data['y'], color='blue', alpha=0.6, label='Data Points')
plt.plot(X, model.predict(data[['X']]), color='red', linewidth=2, label='Regression Line')
plt.xlabel('Independent Variable (X)')
plt.ylabel('Dependent Variable (y)')
plt.title('Linear Regression Fit')
plt.legend()
plt.show()
```

### Step 4: Model Evaluation
Here we evaluate the performance of the linear regression model on the test data. We use the mean squared error (MSE) as a metric to measure the quality of the model's predictions. The MSE is defined as the average of the squared differences between the predicted values and the actual values of the dependent variable (y) in the test data.

This metric gives us an idea of how well the model generalizes to new, unseen data. Lower values of the MSE indicate better performance of the model.

```python
# Step 4: Model evaluation
# Step 4: Model evaluation
from sklearn.metrics import mean_squared_error, r2_score

# Make predictions on the test set
y_pred = model.predict(X_test)

# Calculate Mean Squared Error (MSE) and R-squared (R)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared (R): {r2}")
```

  
### Metrics Explained: 
  The Mean Squared Error (MSE) and R-squared (R) are common metrics used to evaluate the performance of a regression model.
MSE is a measure of the quality of an estimatorit is always non-negative, and the smaller the MSE, the closer we are to finding the line of best fit.
  R-squared is a statistical measure that represents the proportion of the variance for a dependent variable that's explained by an independent variable or variables in a regression model.

Hence, the closer the R-squared value is to 1, the better the model fits the data. In this case, the R-squared value is approximately 0.98, indicating that the model explains 98% of the variance in the dependent variable (y) using the independent variable (X).
