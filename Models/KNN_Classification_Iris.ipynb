```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el dataset de iris
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target

# Aplicar el algoritmo K-means con 3 clusters (ya que sabemos que hay 3 especies)
kmeans = KMeans(n_clusters=3, random_state=0)
df['cluster'] = kmeans.fit_predict(iris.data)

# Visualizar los clusters en 2D usando solo dos de las caracter√≠sticas
plt.figure(figsize=(10, 6))
plt.scatter(df.iloc[:, 0], df.iloc[:, 1], c=df['cluster'], cmap='viridis', label="K-means clusters")
plt.xlabel('Sepal length (cm)')
plt.ylabel('Sepal width (cm)')
plt.title('Clustering de Iris con K-means')
plt.colorbar(label='Cluster')
plt.show()
```

## K-NN

Paso a Paso del Ejemplo de K-nearest neighbors (KNN)
Cargar el Dataset Iris: Vamos a usar las cuatro caracter√≠sticas (longitud y ancho del s√©palo y del p√©talo) como variables independientes, y la especie de la flor como la variable objetivo.

Dividir el Dataset: Dividimos el dataset en conjuntos de entrenamiento y prueba para evaluar la precisi√≥n del modelo.

Configurar el Modelo KNN: Establecemos el n√∫mero de vecinos k (por ejemplo, k=5) para hacer las predicciones. KNN busca los 5 vecinos m√°s cercanos para clasificar cada muestra.

Entrenar y Evaluar el Modelo: Entrenamos el modelo con el conjunto de entrenamiento y luego lo probamos en el conjunto de prueba para ver su precisi√≥n.

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Cargar el dataset de iris
iris = load_iris()
X = iris.data  # Caracter√≠sticas
y = iris.target  # Etiquetas (especies de flores)

# Dividir el dataset en conjunto de entrenamiento y de prueba (70% entrenamiento, 30% prueba)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# Configurar el modelo KNN con k=5 vecinos
knn = KNeighborsClassifier(n_neighbors=5)

# Entrenar el modelo con los datos de entrenamiento
knn.fit(X_train, y_train)

# Hacer predicciones en el conjunto de prueba
y_pred = knn.predict(X_test)

# Calcular la precisi√≥n del modelo
accuracy = accuracy_score(y_test, y_pred)
print("Precisi√≥n del modelo KNN:", accuracy)
```

### Hacemos prediccion con un ejemplo

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

# Cargar el dataset de iris
iris = load_iris()
X = iris.data  # Caracter√≠sticas
y = iris.target  # Etiquetas (especies de flores)

# Dividir el dataset en conjunto de entrenamiento y de prueba (70% entrenamiento, 30% prueba)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# Configurar el modelo KNN con k=5 vecinos
knn = KNeighborsClassifier(n_neighbors=5)

# Entrenar el modelo con los datos de entrenamiento
knn.fit(X_train, y_train)

# Seleccionar una muestra del conjunto de prueba para predecir
sample_index = 0  # Escogemos la primera muestra en el conjunto de prueba
sample_data = X_test[sample_index].reshape(1, -1)  # Tomamos las caracter√≠sticas de esta muestra

# Predicci√≥n de la especie usando el modelo KNN
predicted_class = knn.predict(sample_data)[0]  # Clase predicha
actual_class = y_test[sample_index]  # Clase real

# Obtener los nombres de las especies
predicted_species = iris.target_names[predicted_class]
actual_species = iris.target_names[actual_class]

print("Predicci√≥n de la especie:", predicted_species)
print("Especie real:", actual_species)
```
